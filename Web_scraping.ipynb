### Edible Mushrooms ( unnecessary )

```python
def fetch_wikipedia_urls(url="https://en.wikipedia.org/wiki/Edible_mushroom"):
    headers = {
        "User-Agent": "MushroomScraper/1.0 (contact: example@example.com)"
    }

    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    urls = []

    table = soup.find("table", class_=["wikitable", "sortable", "jquery-tablesorter"])
    if table is None:
        raise ValueError("Nie znaleziono tabeli z grzybami.")

    for tr in table.find_all("tr"):
        tds = tr.find_all("td")

        if len(tds) >= 3:
            third_td = tds[2]
            a = third_td.find("a")

            if not a:
                continue

            if not "/wiki/" in a["href"]:
                continue

            if "redlink=1" in a["href"]:
                continue

            #url = "https://en.wikipedia.org" + a["href"]
            url = a["href"].strip("/wiki/")

            urls.append(url)

    return urls
```

```python
wiki_urls.extend(fetch_wikipedia_urls())
```

### Get Text wikipedia

```python
API_URL = "https://en.wikipedia.org/w/api.php"

HEADERS = {
    "User-Agent": "WikipediaTextExtractor/1.0"
}

def get_wikipedia_text_api(title):
    params = {
        "action": "query",
        "format": "json",
        "titles": title,
        "prop": "extracts",
        "explaintext": True,
        #"exsectionformat": "plain",
        "redirects": 1
    }

    try:
        response = requests.get(API_URL, params=params, headers=HEADERS, timeout=10)
        response.raise_for_status()

        data = response.json()

        if "query" not in data:
            return f"Error: No query results. Response: {data}"

        pages = data["query"]["pages"]
        page = next(iter(pages.values()))

        if "missing" in page:
            return f"Error: Page '{title}' not found"

        #title = page.get("title")
        extract = page.get("extract", "")
        return {title: extract}

    except requests.exceptions.JSONDecodeError as e:
        return f"Error: Invalid JSON response. Status code: {response.status_code}"
    except requests.exceptions.RequestException as e:
        return f"Error: Request failed - {str(e)}"
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
import re

UNWANTED_SECTIONS = {
    "References",
    "External links",
    "See also",
    "Sources",
    "Further reading",
    "Gallery",
    "Works cited"
}

def extract_sections(data):
    result = {}

    for species_name, text in data.items():
        sections = {}

        parts = re.split(r'\n={2,3} ([^=\n]+) ={2,3}\n', text)

        sections["Introduction"] = parts[0].strip()

        for i in range(1, len(parts), 2):
            if i + 1 < len(parts):
                section_name = parts[i].strip()

                if section_name in UNWANTED_SECTIONS:
                    continue

                section_content = parts[i + 1].strip()

                pattern = r'(=+)\s*([^=\n]+?)\s*\1'
                def replace_header(match):
                    title = match.group(2).strip()
                    return title

                section_content = re.sub(pattern, replace_header, section_content).strip()

                if section_content:
                    sections[section_name] = section_content

        result[species_name] = sections

    return result
```

```python
wiki_data = {}
#wiki_urls

for url in mushroom_species:
    text = get_wikipedia_text_api(url)
    if type(text) is str:
        continue
    cleaned = extract_sections(text)
    wiki_data.update(cleaned)
```

```python
with open("Knowledge_base/wikipedia_clean.json", "w", encoding='utf-8') as outfile:
    json.dump(wiki_data, outfile, ensure_ascii=False, indent=2)
```

## [mushroom.world](https://www.mushroom.world/mushrooms/namelist)

```python
url = 'https://www.mushroom.world/mushrooms/namelist'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

urls = ['https://www.mushroom.world' + a['href'] for a in soup.find_all('a', class_='alist-link', href=True)]
```

```python
def fetch_mushroom_info(url):
    r = requests.get(url, timeout=20)
    print(r.raise_for_status())
    soup = BeautifulSoup(r.text, "html.parser")

    name = soup.find('b', class_='mush-captiontext')
    scientific_name = name.find(text=True, recursive=False).strip()
    name_span = soup.find('span', class_='mush-commonname')
    common_name = name_span.get_text(strip=True).strip('()') if name_span else 'Unknown'

    mushroom_data = {}
    mushroom_data['Scientific name'] = scientific_name
    mushroom_data['Common name'] = common_name
    for info in soup.select(".mush-info"):
        label = info.select_one(".mush-labelus")
        text = info.select_one(".mush-textus, .mush-longtextus")
        if label and text:
            key = label.get_text(strip=True)
            value = text.get_text(' ', strip=True)
            mushroom_data[key] = value

    return mushroom_data, scientific_name
```

```python
all_mushrooms_data = {}

for url in urls:
    data, name = fetch_mushroom_info(url)
    all_mushrooms_data[name] = data
```

```python
with open("mushroom_world.json", "w", encoding='utf-8') as outfile:
    json.dump(all_mushrooms_data, outfile, ensure_ascii=False, indent=2)
```

```python
i = 0
for species in mushroom_species:
    if species in all_mushrooms_data:
        i += 1
print(i)
```

## [Mushroom Guide](https://www.wildfooduk.com/mushroom-guide/)

```python
url = 'https://www.wildfooduk.com/mushroom-guide/?_per_page=300'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

urls = []

tds = soup.find_all('td', class_='mushroom-image')
for td in tds:
  url = td.find('a', href=True)['href']
  urls.append(url)
```

```python
all_mushrooms_data = {}

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    edibility = soup.find('div', class_='mush-icon-label').get_text(strip=True)
    description = soup.find('div', class_='entry-content').find('p').get_text(strip=True)
    div = soup.find('div', class_='wp-container-core-group-is-layout-0b691c38')
    table = div.find('table')

    mushroom_content = soup.find('div', id='mushroom-content')

    mushroom_data = {'Description': description}

    for header_td in table.find_all("td", class_="spec-header"):
        key = header_td.get_text(strip=True)

        value_td = header_td.find_next_sibling("td")
        value = value_td.get_text(" ", strip=True) if value_td else ""
        value = value.replace('\xa0', ' ') if value else ""

        if key == 'Mushroom Type':
            mushroom_data['Edibility'] = edibility
            continue

        if key == 'Common Names' and value:
            common_names_dict = {}
            parts = value.split(',')
            for part in parts:
                part = part.strip()
                if '(' in part and ')' in part:
                    name = part[:part.rfind('(')].strip()
                    lang = part[part.rfind('(')+1:part.rfind(')')].strip()
                    if name and lang:
                        common_names_dict[lang] = name
            mushroom_data[key] = common_names_dict if common_names_dict else value
        else:
            mushroom_data[key] = value if value else "(no data)"

    for header_td in mushroom_content.find_all("h2"):
      key = header_td.get_text(strip=True)

      value_td = header_td.find_next_sibling("p")
      value = value_td.get_text(" ", strip=True) if value_td else ""
      value = value.replace('\xa0', ' ') if value else ""

      mushroom_data[key] = value if value else "(no data)"

    all_mushrooms_data[mushroom_data['Scientific Name']] = mushroom_data
```

```python
with open("mushroom_guide.json", "w", encoding='utf-8') as outfile:
    json.dump(all_mushrooms_data, outfile, ensure_ascii=False, indent=2)
```

```python
i = 0
for species in mushroom_species:
    if species in all_mushrooms_data:
        i += 1
print(i)
```
